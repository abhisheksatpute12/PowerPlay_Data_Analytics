# User Behaviour Analysis

```python
sns.countplot(x="Section",data=df3)
```

```python

event_data = df3

# Sorting events by timestamp within each user session
event_data.sort_values(by=['user_id', 'created_at_time'], inplace=True)

# Grouping events by user ID to create separate user sessions
session_groups = event_data.groupby('user_id')

# Dictionary to store sequence frequencies
sequence_counts = {}

# Iterate over each user session group
for user_id, session_data in session_groups:
    
    event_sequence = tuple(session_data['Section'])
    
    # Increment sequence count in dictionary
    sequence_counts[event_sequence] = sequence_counts.get(event_sequence, 0) + 1


```
